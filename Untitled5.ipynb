```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Data
years = np.arange(2014, 2024)
reserves = [310, 345, 370, 395, 430, 470, 505, 540, 570, 600]

fig = plt.figure(figsize=(12, 7))
ax = fig.add_subplot(111, projection='3d')

# Coordinates
x = years
y = np.zeros_like(years)
z = np.zeros_like(years)
dx = np.ones_like(years) * 0.5
dy = np.ones_like(years) * 0.5
dz = reserves

# Gradient color
colors = plt.cm.plasma(np.linspace(0, 1, len(years)))

# Bars
ax.bar3d(x, y, z, dx, dy, dz, color=colors, shade=True)

# Year Labels on top of each bar
for i in range(len(years)):
    ax.text(x[i], y[i], dz[i] + 10,     # Slightly above each bar
            f"{reserves[i]}B",
            color='black', ha='center', fontsize=9, fontweight='bold')

# Trend Line
ax.plot(x, y, dz, color='black', linestyle='--', linewidth=2, label='Growth Trend')

# Titles & Labels
ax.set_title("India's Forex Reserves (20142023)", fontsize=14, fontweight='bold')
ax.set_xlabel('Year')
ax.set_ylabel('')
ax.set_zlabel('Billion USD')

# 3D angle
ax.view_init(elev=20, azim=135)
plt.legend()
plt.tight_layout()
plt.show()
```
